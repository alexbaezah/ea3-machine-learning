```python
# Usando polyfit para una funcion de grado 1 obtener la ecuacion de la recta 
# para gastos_totales = a la sumatoria de youtube,Face,Instagram

# Obtener los datos de las variables predictoras y la variable objetivo
suma = df['Face'] + df['Instagram'] + df['Youtube']
x = suma
y = df['sales']


# Ajustar una regresión lineal de grado 1
coefficients = np.polyfit(x, y, 1)

# Obtener los coeficientes de la ecuación de la recta
a = coefficients[0]
b = coefficients[1]

# Imprimir la ecuación de la recta
print(f"La ecuación de la recta es: sales = {a:.2f} * (Youtube + Face + Instagram) + {b:.2f}")
```

### TEMA A TRATAR

Exploramos **¿Existe una relación entre el gasto publicitario *total* y las *ventas*?**, así como la predicción de las ventas totales para algún valor del gasto total. Ahora queremos expandir esto a **¿Cuál es la relación entre cada canal de publicidad (youTube, face, instagram) y las ventas?*

```python
# Definir el rango de valores de potential_spend
potential_spend = np.linspace(0, 600, 100)

# Calcular las ventas predichas utilizando la ecuación de la recta
predicted_sales = a * potential_spend + b

# Imprimir los resultados
for spend, sales in zip(potential_spend, predicted_sales):
    print(f"Gasto potencial: {spend:.2f}  Ventas predictorias: {sales:.2f}")
```

```python
#Graficar el scaterplot y la recta

# Graficar el scatterplot de los datos
plt.scatter(x, y, color='blue', label='Datos reales')

# Graficar la recta de regresión
plt.plot(x, a*x + b, color='red', label='Recta de regresión')

# Etiquetas de los ejes y título del gráfico
plt.xlabel('Suma de gastos (Youtube + Face + Instagram)')
plt.ylabel('Ventas')
plt.title('Scatterplot y recta de regresión')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
#Predecir las ventas en función de gastos de 170, y 287

# Definir los gastos
gastos_1 = 170
gastos_2 = 287

# Predecir las ventas utilizando la ecuación de la recta de regresión
ventas_1 = a * gastos_1 + b
ventas_2 = a * gastos_2 + b

# Imprimir las predicciones
print(f"Para un gasto de {gastos_1}, la predicción de ventas es: {ventas_1:.2f}")
print(f"Para un gasto de {gastos_2}, la predicción de ventas es: {ventas_2:.2f}")
```



## Evaluación del test model

### Metrics


Las 3 metricas para evaluar regression problems:

**Mean Absolute Error** (MAE) is the mean of the absolute value of the errors:( Calcula el error entre el true value y la prediccion)

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Mean Squared Error** (MSE) es el mean de los squared errors ( para resolver los problemas de larger error)

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Root Mean Squared Error** (RMSE) es el square root del mean de los squared errors:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

Comparing these metrics:

- **MAE** Es facil de comprender es el error AVG
- **MSE** Es mas popular que MAE  ya que hace "punish" a los larger errors, lo cual es util en el mundo real
- **RMSE** Es interpretado en "y" units.

Todas esta funciones minimizan el error

### Evaluacion del modelo:
